## Realizar una función llamada **año_bisiesto**:

- Recibirá un año por parámetro
- Imprimirá “El año año es bisiesto” si el año es bisiesto
- Imprimirá “El año año no es bisiesto” si el año no es bisiesto
- Si se ingresa algo que no sea número debe indicar que se ingrese un número

### Información a tener en cuenta al realizar el entregable:

Se recuerda que los años bisiestos son múltiplos de 4, pero los múltiplos de 100 no lo son, aunque los múltiplos de 400 sí. Estos son algunos ejemplos de posibles respuestas: 2012 es bisiesto, 2010 no es bisiesto, 2000 es bisiesto, 1900 no es bisiesto.

Version *suficiente*

```python
def anio_bisiesto(anio_string):
  anio = anio_string
  while not anio.isnumeric():
    anio = input('\nIngrese un año valido => ')
  
  anio = int(anio)

  if anio%4 == 0 and (anio%100 != 0 or anio%400 == 0):
    print(anio, 'ES bisiesto!')
    return
  print(anio, 'NO es bisiesto!')


anio_bisiesto(input('Ingrese un año => '))
print()
anio_bisiesto(input('Ingrese un año => '))
```

Version Alternativa

```python
# Modularizando codigo
verify_int_input = lambda string_integer:  int(string_integer) if (string_integer.isnumeric() or (len(string_integer) and string_integer[0] == '-' and string_integer[1:].isnumeric())) else verify_int_input(input('Ingreso incorrecto.\nIngrese un numero\n\t=> '))
verify_leap_year = lambda year: year%4 == 0 and (year%100 or year%400 == 0)

def anio_bisiesto(anio):
  es_bisiesto = verify_leap_year(anio)
  print('\n',f'El año {anio} es bisiesto' if es_bisiesto else f'El año {anio} NO es bisiesto')

anio_a_pruebaA = verify_int_input(input('Ingrese un numero equivalente a un año\n\t=> '))
anio_bisiesto(anio_a_pruebaA)

anio_a_pruebaB = verify_int_input(input('Ingrese un numero equivalente a un año\n\t=> '))
anio_bisiesto(anio_a_pruebaB)
```

### Pruebita:
Justificacion de la validacion de numeros del ejercicio anterior mediante la fn *verify_int_input*.
- Se obtendria error al intentar convertir a numero un input vacio ("")
- Se obtendria False al hacer "*numero_negativo*".isnumeric()
- Si intentamos considerar los numeros negativos en un string, se debe verificar que existe el signo "-" procedido de valores numericos
- Al validar el punto anterior mediante slicing, debe validarse previamente el largo de la cadena

<sup><sub>Al final convendria usar expresiones regulares pero estamos aprendiendo y practicando</sub></sup>

```python
import re
number_regex = re.compile(r'^\-?[1-9][0-9]*$')

small_test_values = ['100', '-100', '-', '']

# Variacion no recursiva de verify_int_input, devuelve solamente True/False
is_number = lambda string_integer: string_integer.isnumeric() or (bool(len(string_integer)) and string_integer[0] == '-' and string_integer[1:].isnumeric())

for value in small_test_values:
  print(f"""'{value}'{'' if len(value) else ' (VACIO)'}:
    isnumeric: {value.isnumeric()}
    Regex validation: {bool(re.match(number_regex, value))}
    Own FN validation: {is_number(value)}\n""")
```
